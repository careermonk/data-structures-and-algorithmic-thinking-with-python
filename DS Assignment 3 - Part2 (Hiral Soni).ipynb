```python
#Linked List


import math

class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
    
class LinkedList:
    def __init__(self):
        self.head = None
    
    def addToList(self,data):
        newNode = Node(data)
        
        if self.head == None: #first node
            self.head = newNode
            
        else: #traverse to end of list
            current = self.head
            
            while current.next is not None:
                current = current.next
            current.next = newNode
            
        
    
    def __str__(self):
        if self.head == None:
            return "Empty List"
        current = self.head
        temp = str(current.data)
        while current.next is not None:
            current = current.next
            temp = temp + " -->"+str(current.data)
        return temp
    
    def findcycle(self):
        r = t = self.head
        while r and r.next:
            r = r.next.next
            t = t.next
            if r == t:
                return True
        return False
    
    def startOfCycle(self):
        r = t = self.head
        hascycle = False
        while r and r.next:
            r = r.next.next
            t = t.next
            if r == t:
                hascycle = True
                break
        if hascycle == True:
            r = self.head
            while r!= t:
                r = r.next
                t = t.next
            return r.data
        
    def lengthOfCycle(self):
        c = 0
        r = t = self.head
        while r !=t:
            print('in while')
            r= r.next
            t = t.next
            c +=1
        return c
    
    def findCycleFlyod(self):
        slow=self.head
        fast=self.head
        curr=None
        while slow and fast and fast.next:
            slow=slow.next
            fast=fast.next.next
        
            if slow==fast:
                curr=slow
                break
    
        if curr:
            r=self.head
            c=0
            l=0
            while r:
                l=l+1
                if r==curr and c<1:
                    c=c+1
                    l=0
                elif r==curr and c>=1:
                    break
                
                r=r.next
            return True
        else:
            return False
        

# Find Merge Node between in a given Linked List

if __name__ == '__main__':
    
    l2 = LinkedList()
    l2.addToList(1)
    l2.addToList(4)
    l2.addToList(2)
    l2.addToList(9)
    l2.addToList(15)
    l2.addToList(10)
    l2.addToList(40)
    l2.addToList(20)
    l2.addToList(90)
    l2.addToList(8)

    
    l2.head.next.next.next.next.next.next = l2.head.next.next.next
    
    print("Cycle Found in List ::",l2.findcycle())
    print("Cycle Found using Floyd Method in List ::",l2.findCycleFlyod())
    print("Start of Cycle in List ::",l2.startOfCycle())
    
    
   

```
